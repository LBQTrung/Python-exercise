```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Đọc file dữ liệu:

```python
df = pd.read_csv("StudentsPerformance.csv")
```

Một vài thông tin liên quan của tập dữ liệu:

```python
df.head()
```

```python
df.describe()
```

Lấy mẫu ngẫu nhiên đơn giản:

```python
df.sample(n = 30,
          random_state = 1)
```

**Lấy mẫu ngẫu nhiên phân tầng:**

* Ta tiến hành lấy mẫu theo các group của biến race/ethnicity, Trước tiên ta xem xét biến trên gồm những giá trị nào:

```python
list(df["race/ethnicity"].value_counts().keys())
```

* Vì ta cần chọn ra một mẫu có 30 phần tử nên mỗi group ta sẽ chọn ngẫu nhiên 6 phần tử:

```python
df.groupby('race/ethnicity', group_keys=False).apply(lambda x:x.sample(6))
```
